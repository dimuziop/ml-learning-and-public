```python
def f(x):
    return torch.pow((x - 2.0), 2)
```

## Minimizing funcitons

```python
x = torch.tensor([-3.5], requires_grad = True)
```

```python
x_cur = x.clone()
x_prev = x_cur*100
```

```python
epsilon = 1e-5 # Threshold for the current and previous value to be close enough to stop
eta = 0.1 # learning rate (step magintude)
```

```python
while torch.linalg.norm(x_cur - x_prev) > epsilon:
    x_prev = x_cur.clone()
    value = f(x)
    value.backward()
    x.data -= eta * x.grad

    x.grad.zero_()
    x_cur = x.data

print(x_cur)
    
```

# Optimizing parameters

```python
x_param = torch.nn.Parameter(torch.tensor([-3.5], requires_grad = True))
optimizer = torch.optim.SGD([x_param], lr=eta)
```

```python
for epoch in range(60):
    optimizer.zero_grad()
    loss_incurred = f(x_param)
    loss_incurred.backward()
    optimizer.step()
print(x_param.data)
```

```python
X, y = fetch_openml('mnist_784', version=1, return_X_y=True, as_frame=False) # returns numpy intead dataframe
print(X.shape)
```

```python
print(y)
```

```python
class SimpleDataset(Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y

    def __getitem__(self, index):
        inputs = torch.tensor(self.X[index,:], dtype=torch.float32)
        targets = torch.tensor(int(self.y[index]), dtype=torch.int64)
        return inputs, targets

    def __len__(self):
        return self.X.shape[0]

dataset = SimpleDataset(X, y)
        
```

```python
print("Lenght: ", len(dataset))
example, label = dataset[9]
print("Features:", example.shape)
print("Label of index 0: ", label)
```
