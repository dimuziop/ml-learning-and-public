### Exercise 3

For every power of 2 (i.e., $2^i$) up to $2^{11}$, create a random matrix $X \in \mathbb{R}^{2^i \times 2^i}$ 
(i.e., `X.shape` should give `(2**i, 2**i)`).

Time how long it takes to compute $X \cdot X$ (i.e., `X @ X`) on a CPU and on a GPU, and plot the speedup.

**Question:** For what matrix sizes is the CPU faster than the GPU?

```python
import timeit

cpu_results = np.array([0])
for i in range(1, 11):
    size = 2**i
    x = torch.rand(size, size)
    cpu_results = np.append(cpu_results, timeit.timeit("x@x", globals=globals(), number=100))
    #cpu_result[i] = timeit.timeit("x@x", globals=globals(), number=100)

print(cpu_results)
```

```python
import timeit

device = torch.device("cuda")

gpu_results = np.array([0])
for i in range(1, 11):
    size = 2**i
    x = torch.rand(size, size).to(device)
    gpu_results = np.append(gpu_results, timeit.timeit("x@x", globals=globals(), number=100))
    #cpu_result[i] = timeit.timeit("x@x", globals=globals(), number=100)

print(gpu_results)
```

### Exercise 4

We used PyTorch to find the numeric solution to $f(x) = (x - 2)^2$. 

Write code that finds the solution to:

$$f(x) = \sin(x - 2) \cdot (x + 2)^2 + \sqrt{|\cos(x)|}$$

What answer do you get?

```python
def f(x):
    return torch.sin(x-2) * (x+2)**2 + torch.sqrt(torch.abs(torch.cos(x)))
```

```python
print(f(torch.tensor([2])))
```

```python
x_vals = torch.linspace(-50, 50, 100000)
y_vals = [f(torch.tensor(x)).item() for x in x_vals]
plt.plot(x_vals, y_vals)
plt.grid(True)
plt.show()
```

```python
x = torch.tensor([3.0], requires_grad=True)  # UN solo valor inicial

x_cur = x.clone()
x_prev = x.clone() * 5000

epsilon = 1e-5
eta = 0.1

while torch.linalg.norm(x_cur - x_prev) > epsilon:
    print(x_curr, x_prev)
    x_prev = x_cur.clone()
    value = f(x)
    value.backward()
    x.data -= eta * x.grad

    x.grad.zero_()

    x_cur = x.data

print(f"Convergi a x = {x_cur.item()}")
print(f"f(x) = {f(x).item()}")
```

```python
# Primero graficamos la funcin
x_plot = torch.linspace(-5, 5, 500)
y_plot = [f(torch.tensor([xi])).item() for xi in x_plot]

# Guardamos el recorrido del optimizer
x = torch.tensor([4.0], requires_grad=True)  # empezamos en x=4
eta = 0.1
recorrido_x = [x.item()]
recorrido_y = [f(x).item()]

for i in range(50):
    value = f(x)
    value.backward()
    x.data -= eta * x.grad
    x.grad.zero_()
    recorrido_x.append(x.item())
    recorrido_y.append(f(x).item())

# Graficamos
plt.figure(figsize=(12, 6))
plt.plot(x_plot, y_plot, 'b-', label='$f(x)$', linewidth=2)
plt.plot(recorrido_x, recorrido_y, 'ro-', markersize=8, label='Gradient descent', alpha=0.7)
plt.plot(recorrido_x[0], recorrido_y[0], 'go', markersize=15, label='Inicio')
plt.plot(recorrido_x[-1], recorrido_y[-1], 'r*', markersize=20, label='Final')
plt.grid(True)
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent: el punto "camina" hacia el mnimo')
plt.show()

print(f"Empez en x = {recorrido_x[0]:.4f}")
print(f"Termin en x = {recorrido_x[-1]:.4f}")
```

### Exercise 5

Write a new function that takes two inputs, $x$ and $y$, where:

$$f(x, y) = \exp\left(\frac{\sin(x)^2}{(x - y)^2}\right) + (x - y)^2$$

Use an Optimizer with initial parameter values of $x = 0.2$ and $y = 10$. 

What do they converge to?

```python
def f(x, y):
    return torch.exp((torch.sin(x)**2)/(x-y)**2)+(x - y)**2
```

```python
x_param = torch.nn.Parameter(torch.tensor([0.2], requires_grad=True))
y_param = torch.nn.Parameter(torch.tensor([10.0], requires_grad=True))
eta = 0.1
optimizer = torch.optim.SGD([x_param, y_param], lr=eta)
```

```python
for epoch in range(60):
    optimizer.zero_grad()
    loss_incurred = f(x_param, y_param)
    loss_incurred.backward()
    optimizer.step()

print("x:", x_param.data, "y:", y_param.data,"f(x, y):", f(x_param.data, y_param.data))
    
```

Create a function `libsvm2Dataset` that takes a path to a libsvm dataset file (see https://www.csie.ntu.edu.tw/~cjlin/libsvmtools/datasets/ for many that you can download) and create a new dataset object. 

Check that it is the correct length and that each row has the expected number of features.

```python
class SimpleDataset(Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y

    def __getitem__(self, index):
        inputs = torch.tensor(self.X[index,:], dtype=torch.float32)
        targets = torch.tensor(int(self.y[index]), dtype=torch.int64)
        return inputs, targets

    def __len__(self):
        return self.X.shape[0]
```

```python
from libsvmdata import fetch_libsvm
```

```python
def libsvm2Dataset(path):
    X, y = fetch_libsvm(path)
    return SimpleDataset(X, y)
```

### Exercise 7 (Challenging)

Use NumPy's `memmap` functionality to write the MNIST dataset to disk. Then create a `MemmapedSimpleDataset` that takes the mem-mapped file as input, reading the matrix from disk to create PyTorch tensors in the `__getitem__` method.

**Question:** Why do you think this would be useful?

```python
X, y = fetch_openml('mnist_784', version=1, return_X_y=True, as_frame=False) # returns numpy intead dataframe

targetFile = np.memmap('mnist.npy', dtype='float32', mode='w+', shape=(X.shape[0], 785))
targetFile[:, :784] = X 
targetFile[:, 784] = y 
targetFile.flush()
```

```python
print(targetFile)
```

```python
class MemmapedSimpleDataset(Dataset):
    def __init__(self, filepath, shape):
        self.filepath = filepath
        self.shape = shape
        self.X = None
        self.y = None

    def __getitem__(self, index):
        if (self.X is None or self.y is None):
            self._load_data()
        inputs = torch.tensor(self.X[index,:], dtype=torch.float32)
        targets = torch.tensor(int(self.y[index]), dtype=torch.int64)
        return inputs, targets

    def __len__(self):
        if (self.X is None or self.y is None):
            self._load_data()
        return self.X.shape[0]

    def _load_data(self):
        mem = np.memmap(self.filepath,dtype='float32', mode='r', shape=self.shape)
        self.X = mem[:, :784]
        self.y = mem[:, 784]
```

```python
memmaped_dataset = MemmapedSimpleDataset('array.npy', (70000, 785))
print(memmaped_dataset[5])
```

```python
print("Lenght: ", len(memmaped_dataset))
example, label = memmaped_dataset[5]
print("Features:", example.shape)
print("Label of index 0: ", label)
```
